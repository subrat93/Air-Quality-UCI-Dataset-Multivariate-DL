# Transform Timeseries to Supervised

```python
# frame a sequence as a supervised learning problem
def timeseries_to_supervised(df_local, lag=1):
#     df = pd.DataFrame(data)
    columns = [df_local.shift(i) for i in range(1, lag+1)]
    columns.append(df_local)
    df_local = pd.concat(columns, axis=1)
    df_local.fillna(0, inplace=True)
    return df_local
```

```python
df_supervised = timeseries_to_supervised(df_act, 1)
```

```python
df_supervised.head()
```

# Normalize the dataset

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# scale train and test data to [-1, 1]
def scale(train, test):
    # fit scaler
    scaler = MinMaxScaler(feature_range=(-1, 1))
#     scaler = StandardScaler()
    scaler = scaler.fit(train)
    # transform train
#     train = train.reshape(train.shape[0], train.shape[1])
    train_scaled = scaler.transform(train)
    # transform test
#     test = test.reshape(test.shape[0], test.shape[1])
    test_scaled = scaler.transform(test)
    return scaler, train_scaled, test_scaled
```

```python
# inverse scaling for a forecasted value
def invert_scale(scaler, X, value):
    new_row = [x for x in X] + [val for val in value]
    array = np.array(new_row)
    array = array.reshape(1, len(array))
    inverted = scaler.inverse_transform(array)
    return inverted[0, 9:]
```

```python
scaler, train_scaled, test_scaled = scale(train.values, test)
```

```python
train_scaled
```

# Fit the LSTM Model

```python

from numpy.random import seed
seed(42)
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
# fit an LSTM network to training data
def fit_lstm(train, batch_size, nb_epoch, neurons):
    X, y = train[:, 0:9], train[:, 9:]
    X = X.reshape(X.shape[0], 1, X.shape[1])
    model = Sequential()
    model.add(LSTM(neurons, batch_input_shape=(batch_size, X.shape[1], X.shape[2]), stateful=True))
#     model.add(LSTM(neurons, batch_input_shape=(batch_size, X.shape[1], X.shape[2]), return_sequences=True, stateful=True))
#     model.add(LSTM(neurons, stateful=True))
    model.add(Dense(9))
#     keras.backend.set_epsilon(1)
    model.compile(loss='mean_squared_error', optimizer='adam')
    for i in range(nb_epoch):
        model.fit(X, y, epochs=1, batch_size=batch_size, verbose=1, shuffle=False)
        model.reset_states()
# Below is the code to plot loss vs epoch
#     history = model.fit(train_X, train_y, epochs=50, batch_size=72, validation_data=(test_X, test_y), verbose=2, shuffle=False)
#     # plot history
#     pyplot.plot(history.history['loss'], label='train')
#     pyplot.plot(history.history['val_loss'], label='test')
#     pyplot.legend()
#     pyplot.show()
    return model
```

```python
# fit the model
lstm_model = fit_lstm(train_scaled, 1, 100, 4)
```

```python
# forecast the entire training dataset to build up state for forecasting
train_reshaped = train_scaled[:,0:9].reshape(train_scaled.shape[0], 1, 9)
lstm_model.predict(train_reshaped, batch_size=1)
```

# Forecast LSTM

```python
# make a one-step forecast
def forecast_lstm(model, batch_size, X):
    X = X.reshape(1, 1, len(X))
    yhat = model.predict(X, batch_size=batch_size)
    return yhat[0]
```

```python
# walk-forward validation on the test data
predictions = list()
expected_list = list()
for i in range(len(test_scaled)):
    # make one-step forecast
    X, y = test_scaled[i, 0:9], test_scaled[i, 9:]
    yhat = forecast_lstm(lstm_model, 1, X)
    # invert scaling
    yhat = invert_scale(scaler, X, yhat)
    # invert differencing
    # yhat = inverse_difference(raw_values, yhat, len(test_scaled)+1-i)
    # store forecast
    predictions.append(yhat)
    expected = df.values[len(train) + i]
    expected_list.append(expected)
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    """Calculates MAPE given y_true and y_pred"""
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_pred - y_true) / y_true)) * 100
```

```python
for index, c in enumerate(cols[1:]):
    print(mean_absolute_percentage_error(y_true=df_act[[c]].values[8000:], y_pred=pd.DataFrame(predictions)[index].values))
```
